```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

```

```python
#seed설정.
seed=42
tf.random.set_seed(seed)
np.random.seed(seed)


data_path="./houseprice/train.csv"

read_data=pd.read_csv(data_path)

read_data.info()

```

```python
house_data=np.array(read_data)


# for i in range(len(read_data.columns)):
#     name=read_data.columns[i]
#     name=house_data[:,i]
#     print(name)
    
# for i in range(len(read_data.columns)):
#     column_name = read_data.columns[i] # i번째 열 이름을 가져옴
#     column_data = read_data.iloc[:, i] # i번째 열 데이터를 가져옴
#     print(column_data)
#     #print(column_data) # 열 데이터 출력
#     # 이제 column_name과 column_data 변수를 사용하여 데이터를 저장하면 됩니다.

import numpy as np

data_dict = {} # 빈 딕셔너리 생성

for i in range(len(read_data.columns)):
    column_name = read_data.columns[i] # i번째 열 이름을 가져옴
    column_data = read_data.iloc[:, i].values # i번째 열 데이터를 가져와 numpy 배열로 변환
    data_dict[column_name] = column_data # 딕셔너리에 열 이름과 해당 열의 numpy 배열을 추가


# 딕셔너리에는 각각의 열 이름을 키(key)로, 해당 열 데이터의 numpy 배열을 값(value)으로 가지고 있음
# data_dict를 사용하여 각각의 열 데이터에 접근할 수 있음

# 예시로 첫 번째 열 데이터에 접근


first_column_data = data_dict['Id']
print(first_column_data)


first_column_data = data_dict['LotArea']
print(first_column_data)
```
